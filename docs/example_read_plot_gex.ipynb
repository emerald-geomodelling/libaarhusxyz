# Notebook esample for handling GEX files

this notebook illustrates how to load the content from a skyTEM .gex file into a dictionary using "emerald_aux_data.gex_read" an plot the waveforms

```python
import matplotlib.pyplot as plt
import os.path
from libaarhusxyz import gex_read
```

define input file

```python
input_data_dirname = "./data/"
gex_infile_name = "00473_NOR_SkyTEM304.gex"
#gex_infile_name = "20191126_00473_NOR_SkyTEM304.gex"
gex_infile = os.path.join(input_data_dirname, gex_infile_name)
```

read file into dictionary

```python
gex=gex_read.readfile(gex_infile)
```

Do some plotting of the waveform

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(gex["General"]["WaveformLMPoint"][:,0],gex["General"]["WaveformLMPoint"][:,1], label="LM")
ax.plot(gex["General"]["WaveformHMPoint"][:,0],gex["General"]["WaveformHMPoint"][:,1], label="HM")
ax.grid()
ax.set_title(gex_infile_name)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Amplitude [1]")
#ax.set_ylim([-0.1, 1.1])
plt.legend()
```

do some more detailed plotting of the waveforms

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.semilogx(gex["General"]["WaveformLMPoint"][:,0],gex["General"]["WaveformLMPoint"][:,1], label="LM")
ax.semilogx(gex["General"]["WaveformHMPoint"][:,0],gex["General"]["WaveformHMPoint"][:,1], label="HM")
ax.grid()
ax.set_title(gex_infile_name)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Amplitude [1]")
ax.set_ylim([-0.1, 1.1])
plt.legend()
```
